```python
import pandas as pd
from datetime import datetime
from langdetect import detect

# Function to extract year, month, and day from a date string
def extract_date_info(date_str):
    date_obj = datetime.fromisoformat(date_str[:-5])
    return date_obj.year, date_obj.month, date_obj.day

def detect_language(text):
    try:
        return detect(text)
    except:
        return 'unknown'

# Read CSV into DataFrame
df_2008 = pd.read_csv('2008 Tweets_Metrics.csv')

# Apply extract_date_info function to 'created_at' column and expand into new columns
df_2008[['year', 'month', 'day']] = df_2008['created_at'].apply(lambda x: pd.Series(extract_date_info(x)))

# Detect language of each tweet and create a new column 'lang'
df_2008['lang'] = df_2008['text'].apply(detect_language)

# Rearrange the columns
df_2008 = df_2008[['user_name', 'text', 'lang', 'year', 'month', 'day']]

df_2008
```

```python
import pandas as pd
from datetime import datetime
from langdetect import detect

# Function to extract date info
def extract_date_info(date_str):
    if isinstance(date_str, str):
        date_obj = datetime.fromisoformat(date_str[:-5])
        return date_obj.year, date_obj.month, date_obj.day
    else:
        return None, None, None

# Function to detect the language of a text
def detect_language(text):
    try:
        return detect(text)
    except:
        return 'unknown'

# Read CSV into DataFrame
df_2011 = pd.read_csv('2011 Tweets_Metrics.csv')

# Drop rows with missing 'created_at' values
df_2011 = df_2011.dropna(subset=['created_at'])

# Apply extract_date_info function to 'created_at' column and expand into new columns
df_2011[['year', 'month', 'day']] = df_2011['created_at'].apply(lambda x: pd.Series(extract_date_info(str(x))))

# Detect language of each tweet and create a new column 'lang'
df_2011['lang'] = df_2011['text'].apply(detect_language)

# Rearrange the columns
df_2011 = df_2011[['user_name', 'text', 'lang', 'year', 'month', 'day']]

df_2011
```
